##**Importing Libraries**
---
*   Requests for making HTTP requests
*   BeautifulSoup module from bs4 package for parsing over html content
*   pandas to make dataframe and converting it into suitable format

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

## Pagination through itterative method and storing all soups in a list

```python
soups = []

for i in range(1,18):
  link = "https://www.flipkart.com/search?q=laptops+under+50k&sid=6bo%2Cb5g&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_16_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_16_na_na_na&as-pos=1&as-type=RECENT&suggestionId=laptops+under+50k%7CLaptops&requestId=867e8a6c-e5a7-4fd2-947b-038b6881bd44&as-backfill=on&p%5B%5D=facets.price_range.from%3DMin&p%5B%5D=facets.price_range.to%3D50000&p%5B%5D=facets.rating%255B%255D%3D3%25E2%2598%2585%2B%2526%2Babove&page="+str(i)
  r = requests.get(link)
  soup = BeautifulSoup(r.text,"lxml")
  sec = soup.find("div",class_="_1YokD2 _3Mn1Gg")
  soups.append(sec)
  print(link)
```

## Itterating over soups for divs and over divs for required data

```python
p_name = []
p_price = []
p_descrip = []
ind = 0

for ind in range(len(soups)):
  name = soups[ind].find_all("div",class_ = "_4rR01T")
  price = soups[ind].find_all("div",class_ = "_30jeq3 _1_WHN1")
  descrip = soups[ind].find_all("ul",class_ = "_1xgFaf")
  for i in range(len(name)):
    temp = name[i].text
    p_name.append(temp)
    temp = price[i].text
    p_price.append(temp)
    temp = descrip[i].text
    p_descrip.append(temp)
```

## Cleaning the p_name[] list from extra data

```python
for i in range(len(p_name)):
  p_name[i] = p_name[i].split(" - ")[0]
  print(p_name[i])
```

## Creating another column for brand name

```python
p_brand = []
for i in range(len(p_name)):
  p_brand.append(p_name[i].split()[0])
  print(p_brand[i])
```

## Transforming Data into DataFrame

```python
data = {
    "Brand" : p_brand,
    "Title" : p_name,
    "Price" : p_price,
    "Specs" : p_descrip
}

df = pd.DataFrame(data)
print(df.head())
```

## Converting DataFrame into Excel File

```python
df.to_excel("Flipkart_Laptops_under_50k.xlsx",index="False")
```
